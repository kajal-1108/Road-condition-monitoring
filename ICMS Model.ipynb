```python
results = pd.read_csv("results.csv", header=None)
results = results.sort_values(by=0)
```

```python
test_set = pd.DataFrame({'x':xtest,'y':ytest})
```

```python
test_set['image_name'] = test_set['x'].apply(lambda x: x[-16:-4] + ".jpg")
```

```python
test_set = test_set.merge(results, left_on='image_name', right_on=0)
test_set.head()
```

```python
def check_for_cracks(string):
  cracking = 0
  if  isinstance(string,float):
    return cracking
  else:
    split = string.split()
    for i in range(int(len(split)/5)):
      if int(split[i*5]) == 1 or int(split[i*5]) == 2:
        cracking = 1
  return cracking
```

```python
test_set['predicted_cracking'] = test_set[1].apply(lambda x: check_for_cracks(x))
```

```python
from sklearn import metrics

print(metrics.accuracy_score(test_set['y'],test_set['predicted_cracking']))
print(metrics.confusion_matrix(test_set['y'],test_set['predicted_cracking']))
print(metrics.recall_score(test_set['y'],test_set['predicted_cracking']))
print(metrics.precision_score(test_set['y'],test_set['predicted_cracking']))
print(metrics.f1_score(test_set['y'],test_set['predicted_cracking']))
```
